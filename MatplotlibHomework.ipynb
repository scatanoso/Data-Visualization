# Simple Trend Line

```python
plt.plot(years, world_population, 
    color     = 'black',
    marker    = 'x', 
    linestyle = 'dotted'
)
plt.title("Total World Population by Year")
plt.axis('off')

plt.show()
```

The above graph doesn't give the viewer any useful information beyond the fact that the population is increasing - which is a trend.  There are no axis or labels.  There are no  measurements or times listed.  In relation to Kosslyn's Principles:
1.  The Principle of Relevance - there is too little information presented.
2.  The Principle of Appropriate Knowledge - there is so little conveyed here that you don't know what you don't know.  You see a trend that is increasing, but the only hint to what it is is the title.  The user can tell it is an upward trend, but know no time period or amount.
3.  The Principle of Salience - there is no way to know if there are any differences without a scale.  This could be the increase in population over 1 year, 10 years or 100 years.  It looks very much like a straight line, but there is no scale so even knowing that it is a linear realtionship is not certain.
4.  The Principle of Discriminability - without a scale, there is no way to know when and by how much the population changed.
5.  The Principle of Perceptual Organization - the marks seem to form a line as they are all headed in the same direction so the mind groups them together as a straight line, but the data are discrete points in time.
6.  The Principle of Compatibility - One would assume that these are discrete points over time so a bar graph or scatter plot would be a better representation of the data.  If a trend line was desired to be shown, the line could be fit over the bar/scatter plot to show the trend.
7.  The Principle of Informative Changes - you can't really see what information is being transmitted by the changes as this looks like one line vs data points over a specified time period.  It's difficult to tell what information is being relayed - other than the population is increasing, but by how much or over what time period is a mystery.
8.  The Principle of Capacity Limitations - there are no issues here with capacity limits as there is not much data that is memorable being relayed so there isn't much to be remembered!

The graph of the data on The World Bank's website is much more informative.  It has the y label labeled in Billions with background lines crossing the graph so it is easy to follow across to see the numbers.  They x-axis is labeled with the years.  They have inserted a slider bar so that you can change the range of years being looked at.  Their data points are discrete solid circles connected to the ones next to it by a line.  So as the user hovers over the line, they can see the specific population as of the specific year.  The trend is a straight line as opposed to exponential or other type of line, but with discrete identifiable data points.  The data sources are footnoted so the user knows what data is included in the graph.  There is also a link showing the details of the data and a sub-link showing the metadata that makes up the graph.


# Simple Line Chart

```python
plt.plot(years, high_income_nation_populations, 'g-.', label='High Income')
plt.plot(years,  low_income_nation_populations, 'b--', label='Low Income')
plt.legend(loc=0)

plt.show()
```

This graph is even less informative than the first one.  In relation to Kosslyn's Principles:
1.  The Principle of Relevance - there is too little information presented.  A title is needed to show what the graph is displaying.  As we have the data listed as just numbers, there is no way of knowing which countries the data belong to.  Knowing that and making some sort of map graph may be more appropriate depending on what information is being relayed to the user.
2.  The Principle of Appropriate Knowledge - the user sees 2 lines over time that measure income at 2 different levels, but there is no definition of what the high and low income relate to.
3.  The Principle of Salience - It's obvious that there is a difference between high and low income - but no idea of what high and low income refer to or the magnitude of the income due to a lack of labeling on the y axis.
4.  The Principle of Discriminability - It is easy to see the lines are measuring different things as they are 2 different colors and patterns.
5.  The Principle of Perceptual Organization - Again, it is hard to see the data at discrete points in time as it is being represented as a line.
6.  The Principle of Compatibility - One would assume that these are discrete points over time so a bar graph or scatter plot would be a better representation of the data. 
7.  The Principle of Informative Changes - this may have been better presented as stacked bar charts so you could easily see when the data changes in relation to each other.  It is very difficult so see that as drawn.
8.  The Principle of Capacity Limitations - There are 2 values being displayed. If organized properly, it should be easily remembered.


# Simple Bar Chart

```python
plt.bar(years, world_population)
plt.ylabel("Total World Population")
plt.ylim(min(world_population), max(world_population))

plt.show()
```

This graph is similar to the first one, but still needs to contain more information.  In relation to Kosslyn's Principles:
1.  The Principle of Relevance - there is too little information presented.  A title is needed to show what the graph is displaying. It appears that the title was used as the y axis instead of the y axis being labeled as the units of measure for the population
2.  The Principle of Appropriate Knowledge - You can see that there are specific measurements for discrete years, but it's not clear what the measurements are for.
3.  The Principle of Salience - Due to the use of bars, you can see the differences between the yearly measurements.
4.  The Principle of Discriminability - There is only one thing being measured here.  It is easy to see that it is a different value for each year.
5.  The Principle of Perceptual Organization - There is only one thing being measured and since the x axis is so compact it makes it difficult to see the exact bar for each year; the do seem to blend together with 3 while lines spaced appearing (but not really on the graph) to cut the graph into 4 sections.
6.  The Principle of Compatibility - A bar graph is a good representation of discrete points.
7.  The Principle of Informative Changes - you can see that each line represents a different value, but as they are so close to each other horizontally and vertically it is difficult to see the differences.
8.  The Principle of Capacity Limitations - Only one thing is being presented.  If properly displayed, it should be easily remembered.  Right now I can see that the population is being displayed over time, but I can't see the starting or ending population counts.
